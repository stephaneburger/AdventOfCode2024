```python
ini_secrets = list(map(int, data.split("\n")))
ini_secrets[:5]
```

```python
# Calculate the result of multiplying the secret number by 64. Then, mix this result into the secret number. Finally, prune the secret number.
# Calculate the result of dividing the secret number by 32. Round the result down to the nearest integer. Then, mix this result into the secret number. Finally, prune the secret number.
# Calculate the result of multiplying the secret number by 2048. Then, mix this result into the secret number. Finally, prune the secret number.
def iterate(sn):
    r1 = ((sn*64)^sn) % 16777216
    r2 = (round(int((r1/32))^r1)) % 16777216
    r3 = ((r2 * 2048)^r2) % 16777216
    return r3
```

```python
cumsum = 0
seqs = []
vals = []
for s in ini_secrets[:]:
    pn = s
    seq = [0]
    val = [int(str(s)[-1])]
    pv=int(str(s)[-1])
    sn = s
    for _ in range(2000):
        sn = iterate(sn)
        v = int(str(sn)[-1])
        val.append(v)
        seq.append(v-pv)
        pv = v
    cumsum +=sn
    seqs.append(seq)
    vals.append(val)
cumsum
```

```python
import pandas as pd
seqs = pd.DataFrame(seqs)
vals = pd.DataFrame(vals)
```

```python
from tqdm import tqdm
seqs_vals = {}
for i in tqdm(seqs.index):
    seen = []
    for c in seqs.columns[4:]:
        seq_str = str(seqs.loc[i, c-3:c].tolist())
        v = vals.loc[i, c]
        # print(seq_str)
        # print(vals.loc[i, c])
        if seq_str not in seen:
            if seq_str not in seqs_vals.keys():
                seqs_vals[seq_str] = [v]
                seen.append(seq_str)
            else:
                seqs_vals[seq_str].append(v)
                seen.append(seq_str)
        else:
            if seqs_vals[seq_str][-1]<v:
                seqs_vals[seq_str] = seqs_vals[seq_str][:-1]
                seqs_vals[seq_str].append(v)
seqs_tot = {k:sum(l) for k, l in seqs_vals.items()}
max(t for t in seqs_tot.values())                         
```
