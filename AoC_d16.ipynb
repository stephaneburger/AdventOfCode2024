```python
with open("AoC_D16_data.txt", "r") as f:
    data = f.read()
```

```python
maze = pd.DataFrame([list(line) for line in data.split("\n")])
ini_pos = maze.loc[(maze=="S").any(axis=1), (maze=="S").any()]
ini_pos = (ini_pos.columns[0], ini_pos.index[0])
ini_dir = ">"
end_pos = maze.loc[(maze=="E").any(axis=1), (maze=="E").any()]
end_pos = (end_pos.columns[0], end_pos.index[0])
print("ini_pos:", ini_pos, "end_pos:", end_pos)
ini_score_map = pd.DataFrame(data=None, columns=maze.columns, index = maze.index)
ini_score_map.loc[ini_pos[1], ini_pos[0]] = str({">":0})
maze
```

```python
def one_step(maze, score_map, pos, direction):
    x = pos[0]
    y = pos[1]
    # print("start_pos:", pos, score_map.loc[pos[1], pos[0]])
    score = eval(score_map.loc[pos[1], pos[0]])[direction]
    is_front = []
    # outcome: 1 move straight 
    if direction == ">":
        next_pos = (x+1, y)
    elif direction == "v":
        next_pos = (x, y+1)
    elif direction == "<":
        next_pos = (x-1, y)
    elif direction == "^":
        next_pos = (x, y-1)
    if maze.loc[next_pos[1], next_pos[0]] == "#":
        o1 = None
        # print("o1", "blocked")
    else:
        o1 = {direction:score+1}
        if pd.isna(score_map.loc[next_pos[1], next_pos[0]]):
            score_map.loc[next_pos[1], next_pos[0]] = str(o1)
            is_front.append((next_pos, o1))
            # print("o1", "first time", next_pos, o1)
        else:
            nps = eval(score_map.loc[next_pos[1], next_pos[0]])
            if direction not in nps.keys():
                nps.update(o1)
                is_front.append((next_pos, o1))
                # print("o1", "first this dir",next_pos,  nps)
            elif nps[direction]>o1[direction]:
                # score_map.loc[next_pos[1], next_pos[0]][direction] = o1[direction]
                # print("o1", "better score", next_pos, nps, o1)
                nps[direction] = o1[direction]
                is_front.append((next_pos, o1))
            # else:
                # print("o1", "worse score", next_pos, nps, o1)
            score_map.loc[next_pos[1], next_pos[0]] = str(nps)
            
    # outcome 2: Turn Right:
    if direction == ">":
        next_dir = "v"
    elif direction == "v":
        next_dir = "<"
    elif direction == "<":
        next_dir = "^"
    elif direction == "^":
        next_dir = ">"
    o2 = {next_dir:score+1000}
    ps = eval(score_map.loc[pos[1], pos[0]])
    
    if next_dir not in ps.keys():
        ps.update(o2)
        is_front.append((pos, o2))
        # print("o2 first this dir", pos, ps, o2)
    elif ps[next_dir]>o2[next_dir]:
        # print("o2 better this dir", pos, ps, o2)
        ps[next_dir] = o2[next_dir]
        is_front.append((pos, o2))
    # else:
        # print("o2", "worse score", pos, ps, o2)

    score_map.loc[pos[1], pos[0]] = str(ps)
    # outcome 3: Turn Left:
    if direction == ">":
        next_dir = "^"
    elif direction == "v":
        next_dir = ">"
    elif direction == "<":
        next_dir = "v"
    elif direction == "^":
        next_dir = "<"
    o3 = {next_dir:score+1000}
    ps = eval(score_map.loc[pos[1], pos[0]])
    if next_dir not in ps.keys():
        ps.update(o3)
        is_front.append((pos, o3))
    elif ps[next_dir]>o3[next_dir]:
        ps[next_dir] = o3[next_dir]
        is_front.append((pos, o3))
    score_map.loc[pos[1], pos[0]] = str(ps)
    return score_map, is_front
```

```python
score_map = ini_score_map.copy()
pos = ini_pos
direction = ini_dir
is_front_tmp = [((ini_pos), {">": 0})]
next_is_front = []
for _ in range(1000):
    # print("#####", _)
    for front in is_front_tmp:
        # print(front)
        pos = front[0]
        direction = list(front[1].keys())[0]
        score_map, is_front = one_step(maze, score_map, pos, direction)
        next_is_front.extend(is_front)
    # print(next_is_front)
    is_front_tmp = next_is_front.copy()
    next_is_front = []
score_map
```
