```python
with open("AoC_D20_data.txt", "r") as f:
    data = f.read()
```

```python
import pandas as pd
race_map = pd.DataFrame([list(l) for l in data.split('\n')])
ini_picos = pd.DataFrame(data=None, columns = race_map.columns, index=race_map.index)
start = race_map.columns[(race_map=="S").any(axis=0)][0], race_map.index[(race_map=="S").any(axis=1)][0]
end = race_map.columns[(race_map=="E").any(axis=0)][0], race_map.index[(race_map=="E").any(axis=1)][0]
ini_picos.loc[start[1], start[0]] = 0
print(start, end)
race_map
```

```python
directions = [[0, 1], [1, 0], [0, -1], [-1, 0]] #[dy, dx]
pos = start
picos = ini_picos.copy()
p=1
while pos != end:
    for dy, dx in directions:
        # print(pos, dy, dx, race_map.loc[pos[1]+dy, pos[0]+dx], picos.loc[pos[1]+dy, pos[0]+dx])
        if (race_map.loc[pos[1]+dy, pos[0]+dx]!="#") & (pd.isna(picos.loc[pos[1]+dy, pos[0]+dx])):
            picos.loc[pos[1]+dy, pos[0]+dx] = p
            p +=1
            pos = (pos[0]+dx, pos[1]+dy)
            continue

# PART 1
directions = [[0, 2], [2, 0], [0, -2], [-2, 0]] #[dy, dx]  
# PART 2
p2_range = 6
p2_range = 20
directions = [[[dx, dy] for dx in range(-p2_range, p2_range+1)] for dy in range(-p2_range, p2_range+1)] 
directions = sorted([item for row in directions for item in row])

cheat_opportunity = []
for x in picos.columns[1:-1]:
    for y in picos.index[1:-1]:
        if pd.notna(picos.loc[y, x]):
            for dy, dx in directions:
                if (0< x+dx <len(picos.columns)) & (0 < y+dy < len(picos.index)):
                    if pd.notna(picos.loc[y+dy, x+dx]):
                        if picos.loc[y+dy, x+dx] > picos.loc[y, x] + abs(dx) + abs(dy):
                            cheat_opportunity.append(picos.loc[y+dy, x+dx]-picos.loc[y, x] -(abs(dx) + abs(dy)) )
```

```python
# for example
thresh = 50
len([co for co in cheat_opportunity if co>=thresh])
```
