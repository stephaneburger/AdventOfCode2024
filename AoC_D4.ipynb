```python
with open("AoC_D4_data.txt", "r") as f:
    data = f.read()
```

```python
import pandas as pd
ar = pd.DataFrame([list(l) for l in data.split("\n")])
ar
```

```python
padded_ar = ar.copy()
padded_ar.loc[len(padded_ar), :] = "Z"
padded_ar.loc[len(padded_ar), :] = "Z"
padded_ar.loc[len(padded_ar), :] = "Z"
padded_ar.loc[len(padded_ar), :] = "Z"
padded_ar.loc[:, len(padded_ar.columns)] = "Z"
padded_ar.loc[:, len(padded_ar.columns)] = "Z"
padded_ar.loc[:, len(padded_ar.columns)] = "Z"
padded_ar.loc[:, len(padded_ar.columns)] = "Z"
# padded_ar
```

```python
# PART 1
def xmas_count(x):
    c = 0
    if x.iloc[0, :].tolist() == list("XMAS"):
        c += 1
    if x.iloc[0, :].tolist() == list("SAMX"):
        c += 1
    if x.iloc[:, 0].tolist() == list("XMAS"):
        c += 1
    if x.iloc[:, 0].tolist() == list("SAMX"):
        c += 1
    if [x.iloc[0, 0], x.iloc[1, 1], x.iloc[2, 2], x.iloc[3, 3]] == list("XMAS"):
        c += 1
    if [x.iloc[0, 0], x.iloc[1, 1], x.iloc[2, 2], x.iloc[3, 3]] == list("SAMX"):
        c += 1
    if [x.iloc[3, 0], x.iloc[2, 1], x.iloc[1, 2], x.iloc[0, 3]] == list("XMAS"):
        c += 1
    if [x.iloc[3, 0], x.iloc[2, 1], x.iloc[1, 2], x.iloc[0, 3]] == list("SAMX"):
        c += 1
    return c

xmax_count_df = pd.DataFrame(columns=ar.columns, index = ar.index)
for i in xmax_count_df.index:
    for j in xmax_count_df.columns:
        window = padded_ar.loc[i:i+3, j:j+3]
        xmax_count_df.loc[i, j] = xmas_count(window)

print(xmax_count_df.sum().sum())
xmax_count_df
```

```python
# PART 2
def xmas_count2(x):
    c = 0
    if ([x.iloc[0, 0], x.iloc[1, 1], x.iloc[2, 2]] == list("MAS")) & ([x.iloc[2, 0], x.iloc[1, 1], x.iloc[0, 2]] ==list("MAS")) :
        c += 1
    elif ([x.iloc[0, 0], x.iloc[1, 1], x.iloc[2, 2]] == list("MAS")) & ([x.iloc[2, 0], x.iloc[1, 1], x.iloc[0, 2]] ==list("SAM")) :
        c += 1
    elif ([x.iloc[0, 0], x.iloc[1, 1], x.iloc[2, 2]] == list("SAM")) & ([x.iloc[2, 0], x.iloc[1, 1], x.iloc[0, 2]] ==list("MAS")) :
        c += 1
    elif ([x.iloc[0, 0], x.iloc[1, 1], x.iloc[2, 2]] == list("SAM")) & ([x.iloc[2, 0], x.iloc[1, 1], x.iloc[0, 2]] ==list("SAM")) :
        c += 1
    return c

xmax_count_df = pd.DataFrame(columns=ar.columns, index = ar.index)
for i in xmax_count_df.index:
    for j in xmax_count_df.columns:
        window = padded_ar.loc[i:i+2, j:j+2]
        xmax_count_df.loc[i, j] = xmas_count2(window)
print(xmax_count_df.sum().sum())
xmax_count_df
```
