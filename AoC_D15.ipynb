```python
def move(grid, pos, direction):
    # ---> 10 if grid.loc[next_pos[1], next_pos[0]] == "#":

    if direction == "^":
        next_pos = (pos[0], pos[1]-1)
    elif direction == ">":
        next_pos = (pos[0]+1, pos[1])
    elif direction == "v":
        next_pos = (pos[0], pos[1]+1)
    elif direction == "<":
        next_pos = (pos[0]-1, pos[1])
    
    if grid.loc[next_pos[1], next_pos[0]] == "#":
        return grid, pos
    elif grid.loc[next_pos[1], next_pos[0]] == ".":
        return grid, next_pos
    elif grid.loc[next_pos[1], next_pos[0]] == "O":
        # pushing
        if direction == "^":
            row = grid.loc[:next_pos[1], next_pos[0]]
            if (row==".").sum()==0:
                return grid, pos
            else:
                spacey = row.loc[row=="."].index[-1]
                if (row.loc[spacey:]=="#").any():
                    return grid, pos
                grid.loc[spacey, next_pos[0]] = "O"
                grid.loc[next_pos[1], next_pos[0]] = "."
                return grid, next_pos
        elif direction == ">":
            row = grid.loc[next_pos[1], next_pos[0]:]
            # print(row)
            if (row==".").sum()==0:
                return grid, pos
            else:
                spacex = row.loc[row=="."].index[0]
                # print(spacex)
                # print(row.loc[:spacex])
                if (row.loc[:spacex]=="#").any():
                    return grid, pos
                grid.loc[next_pos[1], spacex]  = "O"
                grid.loc[next_pos[1], next_pos[0]] = "."
                return grid, next_pos
        elif direction == "v":
            row = grid.loc[next_pos[1]:, next_pos[0]]
            if (row==".").sum()==0:
                return grid, pos
            else:
                spacey = row.loc[row=="."].index[0]
                if (row.loc[:spacey]=="#").any():
                    return grid, pos
                grid.loc[spacey, next_pos[0]] = "O"
                grid.loc[next_pos[1], next_pos[0]] = "."
                return grid, next_pos
        elif direction == "<":
            row = grid.loc[next_pos[1], :next_pos[0]]
            if (row==".").sum()==0:
                return grid, pos
            else:
                spacex = row.loc[row=="."].index[-1]
                if (row.loc[spacex:]=="#").any():
                    return grid, pos
                grid.loc[next_pos[1], spacex]  = "O"
                grid.loc[next_pos[1], next_pos[0]] = "."
                return grid, next_pos
            
def plot_grid(grid, pos):
    gridplot = grid.copy()
    gridplot.loc[pos[1], pos[0]]="@"
    print(gridplot)
    
def caclulate_gps_sum(grid, part2=False):
    crit = "[" if part2 else "O"
    s = 0
    for x in grid.columns:
        for y in grid.index:
            if grid.loc[y, x]== crit:
                s += 100*y + x
    return s
```

```python
# def box_can_move(lpos, grid, direction, box_can_move_set):
#     assert grid.loc[lpos[1], lpos[0]]=="["
#     new_box_can_move = set()
#     if direction == "^":
#         d = -1
#     else:
#         d = +1
#     if (grid.loc[lpos[1]+d, lpos[0]]==".")&(grid.loc[lpos[1]+d, lpos[0]+1]=="."):
#         box_can_move_bool = True 
#     elif (grid.loc[lpos[1]+d, lpos[0]]=="#")|(grid.loc[lpos[1]+d, lpos[0]+1]=="#"):
#         box_can_move_bool = False
#     elif (grid.loc[lpos[1]+d, lpos[0]]=="[")&(grid.loc[lpos[1]+d, lpos[0]+1]=="]"):
#         box_can_move_bool, new_box_can_move = box_can_move((lpos[0], lpos[1]-1), grid, direction, box_can_move_set)
#     elif (grid.loc[lpos[1]+d, lpos[0]]=="]")&(grid.loc[lpos[1]+d, lpos[0]+1]=="."):
#         box_can_move_bool, new_box_can_move = box_can_move((lpos[0]-1, lpos[1]+d), grid, direction, box_can_move_set)
#     elif (grid.loc[lpos[1]+d, lpos[0]]==".")&(grid.loc[lpos[1]+d, lpos[0]+1]=="["):
#         box_can_move_bool, new_box_can_move = box_can_move((lpos[0]+1, lpos[1]+d), grid, direction, box_can_move_set)
#     elif (grid.loc[lpos[1]+d, lpos[0]]=="]")&(grid.loc[lpos[1]+d, lpos[0]+1]=="["):
#         box_can_move_bool_1, new_box_can_move_1 = box_can_move((lpos[0]-1, lpos[1]+d), grid, direction, box_can_move_set)
#         box_can_move_bool_2, new_box_can_move_2 = box_can_move((lpos[0]+1, lpos[1]+d), grid, direction, box_can_move_set)
#         box_can_move_bool = box_can_move_bool_1 & box_can_move_bool_2 
#         new_box_can_move = new_box_can_move_1.union(new_box_can_move_2)
#     box_can_move_set = box_can_move_set.union(new_box_can_move)
#     if box_can_move_bool:
#         box_can_move = box_can_move.union(new_box_can_move.union(set([lpos])))
#     print(lpos, box_can_move_bool, box_can_move_set)
#     return box_can_move_bool, box_can_move_set
global BOX_TO_MOVE
BOX_TO_MOVE = set()
def box_can_move(lpos, grid, direction):
    assert grid.loc[lpos[1], lpos[0]]=="["
    if direction == "^":
        d = -1
    else:
        d = +1
    if (grid.loc[lpos[1]+d, lpos[0]]==".")&(grid.loc[lpos[1]+d, lpos[0]+1]=="."):
        box_can_move_bool = True
        BOX_TO_MOVE.add((True, lpos))
    elif (grid.loc[lpos[1]+d, lpos[0]]=="#")|(grid.loc[lpos[1]+d, lpos[0]+1]=="#"):
        box_can_move_bool = False
        BOX_TO_MOVE.add((False, lpos))
    elif (grid.loc[lpos[1]+d, lpos[0]]=="[")&(grid.loc[lpos[1]+d, lpos[0]+1]=="]"):
        box_can_move_bool = box_can_move((lpos[0], lpos[1]+d), grid, direction)
        BOX_TO_MOVE.add((box_can_move_bool, lpos))
    elif (grid.loc[lpos[1]+d, lpos[0]]=="]")&(grid.loc[lpos[1]+d, lpos[0]+1]=="."):
        box_can_move_bool = box_can_move((lpos[0]-1, lpos[1]+d), grid, direction)
        BOX_TO_MOVE.add((box_can_move_bool, lpos))
    elif (grid.loc[lpos[1]+d, lpos[0]]==".")&(grid.loc[lpos[1]+d, lpos[0]+1]=="["):
        box_can_move_bool = box_can_move((lpos[0]+1, lpos[1]+d), grid, direction)
        BOX_TO_MOVE.add((box_can_move_bool, lpos))
    elif (grid.loc[lpos[1]+d, lpos[0]]=="]")&(grid.loc[lpos[1]+d, lpos[0]+1]=="["):
        box_can_move_bool_1 = box_can_move((lpos[0]-1, lpos[1]+d), grid, direction)
        box_can_move_bool_2 = box_can_move((lpos[0]+1, lpos[1]+d), grid, direction)
        box_can_move_bool = box_can_move_bool_1 & box_can_move_bool_2 
        BOX_TO_MOVE.add((box_can_move_bool, lpos))
    else:
        print("ERROR")
        print(lpos, direction) 
        print(grid)
        raise
    # print("box_can_move", lpos, box_can_move_bool, BOX_TO_MOVE)
    return box_can_move_bool
```

```python
def move_boxes(lpos_set, grid, direction):
    # print("boxes to move",lpos_set)
    if direction == "^":
        d = -1
    else:
        d = +1
    pos_to_mov = list(lpos_set) + [(x+1, y) for x, y in lpos_set]
    # print(pos_to_mov)
    destination_pos = [(x, y+d) for x, y in pos_to_mov]
    # print(destination_pos)
    pos_to_replace_with_dot = [(x, y) for x, y in pos_to_mov if (x, y) not in destination_pos]
    moves = {(x, y+d):grid.loc[y, x] for x, y in pos_to_mov}
    # print(pos_to_replace_with_dot)
    # for pos in pos_to_mov:
    #     grid.loc[pos[1]+d, pos[0]] = grid.loc[pos[1], pos[0]]
    for pos, char in moves.items():
        grid.loc[pos[1], pos[0]] = char
    for pos in pos_to_replace_with_dot:
        grid.loc[pos[1], pos[0]] = "."
    return grid
    
        
```

```python
# global box_can_move_set
# box_can_move_set = set()

def move_2(grid, pos, direction):
    # print(grid, pos, direction)
    if direction == "^":
        next_pos = (pos[0], pos[1]-1)
    elif direction == ">":
        next_pos = (pos[0]+1, pos[1])
    elif direction == "v":
        next_pos = (pos[0], pos[1]+1)
    elif direction == "<":
        next_pos = (pos[0]-1, pos[1])
    # print(pos, next_pos)
    if grid.loc[next_pos[1], next_pos[0]] == "#":
        return grid, pos
    elif grid.loc[next_pos[1], next_pos[0]] == ".":
        return grid, next_pos
    elif grid.loc[next_pos[1], next_pos[0]] in ["[", "]"]:
        # pushing
        # print(direction)
        if direction in ["^", "v"]:
            # row = grid.loc[:next_pos[1], next_pos[0]]
            # box_can_move_set = set()
            global BOX_TO_MOVE
            BOX_TO_MOVE = set()
            if grid.loc[next_pos[1], next_pos[0]] == "[":
                box_can_move(next_pos, grid, direction)
            else:
                box_can_move((next_pos[0]-1, next_pos[1]), grid, direction)
            # print(BOX_TO_MOVE)
            if all([b[0] for b in BOX_TO_MOVE]) & (len([b[1] for b in BOX_TO_MOVE])>0):
                # return move_boxes(box_can_move_set, grid, direction), next_pos
                return move_boxes([b[1] for b in BOX_TO_MOVE], grid, direction), next_pos
            else:
                return grid, pos
        elif direction == ">":
            row = grid.loc[next_pos[1], next_pos[0]:]
            # print(row)
            if (row==".").sum()==0:
                return grid, pos
            else:
                spacex = row.loc[row=="."].index[0]
                # print(spacex)
                # print(row.loc[:spacex])
                if (row.loc[:spacex]=="#").any():
                    return grid, pos
                grid.loc[next_pos[1], next_pos[0]+1:spacex]  = grid.loc[next_pos[1], next_pos[0]:spacex-1].tolist()
                grid.loc[next_pos[1], next_pos[0]] = "."
                return grid, next_pos
        elif direction == "<":
            row = grid.loc[next_pos[1], :next_pos[0]]
            if (row==".").sum()==0:
                return grid, pos
            else:
                spacex = row.loc[row=="."].index[-1]
                if (row.loc[spacex:]=="#").any():
                    return grid, pos
                grid.loc[next_pos[1], spacex:next_pos[0]-1]  = grid.loc[next_pos[1], spacex+1:next_pos[0]].tolist()
                grid.loc[next_pos[1], next_pos[0]] = "."
                return grid, next_pos
```
