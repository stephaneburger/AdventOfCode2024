```python
with open("AoC_D5_data.txt", "r") as f:
    data = f.read()
```

```python
import pandas as pd
rules = pd.DataFrame([[int(v) for v in rule.split("|")] for rule in data.split("\n\n")[0].split("\n")], columns=["before", "after"])
rules[:]
```

```python
# PART 1
updates = [[int(p) for p in update.split(",")] for update in data.split("\n\n")[1].split("\n")]
def check_page_valid(page_number, page_list, rules):
    pages_before = page_list[:page_number]
    pages_after = page_list[page_number+1:]
    page = page_list[page_number]
    must_be_before = rules.loc[rules["after"]==page, "before"].tolist()
    must_be_after = rules.loc[rules["before"]==page, "after"].tolist()
    if len(set(pages_after).intersection(set(must_be_before)))>0:
        return "push_forward"
    if len(set(pages_before).intersection(set(must_be_after)))>0:
        return "push_backward"
    return "good"

valid_middles = []
for update in updates:
    if all([check_page_valid(i, update, rules)=="good" for i in range(len(update))]):
        valid_middles.append(update[int((len(update)-1)/2)])
sum(valid_middles)
```

```python
# PART 2
updates = [[int(p) for p in update.split(",")] for update in data.split("\n\n")[1].split("\n")]
def move(old_index, update, direction):
    # print("move", update[old_index], direction)
    page = update.pop(old_index)
    if direction == "forward":
        new_position = old_index + 1
    elif direction == "backward":
        new_position = old_index - 1
    return update.insert(new_position, page)

not_valid = []
for update in updates:
    if not all([check_page_valid(i, update, rules)=="good" for i in range(len(update))]):
        not_valid.append(update)
print(len(not_valid), " not valid updates")
print("   ")
        
fixed = []
for update in not_valid:
    print("##### BAD:", update)
    is_not_valid = True
    elements = update.copy()
    msg = ""
    while is_not_valid:
        for e in elements:
            for attempt in range(len(update)):
                i = update.index(e)
                situation = check_page_valid(i, update, rules)
                if situation == "good":
                    # msg += f"{e} is ok; "
                    break
                elif situation == "push_forward":
                    msg += f"{e} forward; "
                    move(i, update, "forward")
                # not needed as we start pushing forward from left
                # elif situation == "push_backward": 
                #     msg += f"{e} backward; "
                #     move(i, update, "backward")
            if all([check_page_valid(i, update, rules)=="good" for i in range(len(update))]):
                fixed.append(update)
                is_not_valid = False
                break
            
    # print(msg)
    print("#-> FIXED:", update)
    print(" ")
valid_middles = [f[int((len(f)-1)/2)] for f in fixed]
sum(valid_middles)
```
