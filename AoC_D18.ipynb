```python
bytes_bombs = [[int(v) for v in line.split(",")] for line in data.split("\n")]
bytes_bombs
mem = pd.DataFrame(data='.', columns=range(width), index=range(height))
ini_step_count = pd.DataFrame(data=9999, columns=mem.columns, index = mem.index)
ini_step_count.loc[0, 0] = 0
```

```python
ini_step_count
```

```python
def one_step(mem, step_count, pos):
    x = pos[0]
    y = pos[1]
    is_front = []
    steps = step_count.loc[pos[1], pos[0]]
    for up, right in [[-1, 0], [0, -1], [1, 0], [0, 1]]:
        if (pos[1]+up in mem.index) & (pos[0]+right in mem.columns):
            if (mem.loc[pos[1]+up, pos[0]+right] != "#") & (steps+1 < step_count.loc[pos[1]+up,pos[0]+right]):
                step_count.loc[pos[1]+up, pos[0]+right] = steps+1
                is_front.append((pos[0]+right, pos[1]+up))
    return step_count, is_front
```

```python
for x, y in bytes_bombs[:bytes_falling]:
    mem.loc[y, x]="#"
mem.iloc[:20, :20]

step_count = ini_step_count.copy()
pos = (0, 0)
is_front_tmp = [(0,0)]
next_is_front = []
while len(is_front_tmp)>0:
    for pos in is_front_tmp:
        step_count, is_front = one_step(mem, step_count, pos)
        next_is_front.extend(is_front)
    is_front_tmp = next_is_front.copy()
    next_is_front = []
print("min steps:", step_count.iloc[-1, -1])
step_count
```
