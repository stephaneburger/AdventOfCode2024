```python
with open("AoC_D8_data.txt", "r") as f:
    data = f.read()
```

```python
import pandas as pd
antennas = pd.DataFrame([list(row) for row in data.split("\n")])
antennas
```

```python
antenna_types = []
for c in antennas.columns:
    antenna_types += antennas[c].replace({".":None}).dropna().unique().tolist()
antenna_types = list(set(antenna_types))
antenna_types
```

```python
def get_all_antennas_pairs(antennas, antenna_type):
    pos = []
    for x in antennas.columns:
        ys = antennas.loc[antennas[x]==antenna_type, x].index.tolist()
        pos += [(x, y) for y in ys]
    return [(p1, p2) for idx, p1 in enumerate(pos) for p2 in pos[idx + 1:]]
get_all_antennas_pairs(antennas, "A")
```

```python
antinodes = pd.DataFrame(data = '', columns=antennas.columns, index = antennas.index)
```

```python
#part 1
for antenna_type in antenna_types:
    # print(antenna_type)
    pairs = get_all_antennas_pairs(antennas, antenna_type)
    for pair in pairs:
        # print(pair)
        a1 = pair[0]
        a2 = pair[1]
        an1 = (a1[0] - (a2[0]-a1[0]), a1[1] - (a2[1]-a1[1]))
        an2 = (a2[0] + (a2[0]-a1[0]), a2[1] + (a2[1]-a1[1]))
        if (0 <= an1[0] < len(antennas.columns)) & (0 <= an1[1] < len(antennas.index)):
            antinodes.loc[an1[1], an1[0]] = "#"
        if (0 <= an2[0] < len(antennas.columns)) & (0 <= an2[1] < len(antennas.index)):
            antinodes.loc[an2[1], an2[0]] = "#"
print((antinodes=="#").sum().sum())
antinodes
    
```

```python
# part 2
for antenna_type in antenna_types:
    # print(antenna_type)
    pairs = get_all_antennas_pairs(antennas, antenna_type)
    for pair in pairs:
        # print(pair)
        a1 = pair[0]
        a2 = pair[1]
        for i in range(len(antennas.columns)+1):
            an1 = (a1[0] - i*(a2[0]-a1[0]), a1[1] - i*(a2[1]-a1[1]))
            an2 = (a2[0] + i*(a2[0]-a1[0]), a2[1] + i*(a2[1]-a1[1]))
            if (0 <= an1[0] < len(antennas.columns)) & (0 <= an1[1] < len(antennas.index)):
                antinodes.loc[an1[1], an1[0]] = "#"
            if (0 <= an2[0] < len(antennas.columns)) & (0 <= an2[1] < len(antennas.index)):
                antinodes.loc[an2[1], an2[0]] = "#"
print((antinodes=="#").sum().sum())
antinodes
```
